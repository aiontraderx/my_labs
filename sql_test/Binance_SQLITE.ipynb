```python
def _first_install_db(symbols,start_dt:dt,end_dt =None,logs:bool=False):
#     start_dt= dt(2018,1,1)
    if end_dt==None :
        end_dt  = start_dt +timedelta(days=7)
    end_ts= int(end_dt.timestamp() *1000)
    print(f'{symbols} ',start_dt, ' ::: ' ,end_dt)
    print('='*100)
    res = client.get_historical_klines(symbols,'1m', start_str= int(start_dt.timestamp() *1000),\
                                       end_str= end_ts)

    while res ==[]:
        # shift dt 
        start_dt = end_dt +timedelta(days=7)
        end_dt  = start_dt +timedelta(days=7)
        if logs ==True:
            print(f'Search First Days {start_dt} : {end_dt}')

#         print('TRY ',start_dt, ' ::: ' ,end_dt)

        res = client.get_historical_klines('SOLUSDT','1m', start_str= int(start_dt.timestamp() *1000),\
                                       end_str= int(end_dt.timestamp() *1000))
    if logs ==True:
        print(f'SUCCESS LOAD DATA {symbols} {start_dt} to {end_dt}')
    cols= ['date','open','high','low','close','volume','close_time','quote_vol','num_trades','taker_base','taker_quote','ignore']
    df = pd.DataFrame(res ,columns=cols)
    df= df.loc[:,['date','open','high','low','close','volume']]
    df['date'] = pd.to_datetime(df['date'],unit='ms')
    return df



start_dt= dt(2020,1,1)
# df = _first_install_db('LUNAUSDT',start_dt=start_dt,logs=True)
```

```python

def binance_ohlc(symbol:'str',lookback:int=1,since=None,end=None ):
    t = dt.today() - timedelta(days= lookback) # today 2020 1 8
    if since == None:
        res = client.get_historical_klines(symbol,'1m', int(t.timestamp() *1000))
    else:
        res = client.get_historical_klines(symbol,'1m', since,end)
    
    cols= ['date','open','high','low','close','volume','close_time','quote_vol','num_trades','taker_base','taker_quote','ignore']
    df = pd.DataFrame(res ,columns=cols)
    df= df.loc[:,['date','open','high','low','close','volume']]
    df['date'] = pd.to_datetime(df['date'],unit='ms')
    return df

def binance_update(symbol:list,end_dt=None):
    if isinstance(symbol,str):
        symbol = [symbol]
    for s in symbol:

        df = pd.read_sql(s,engine).set_index('date')
        last_dt = df.tail(1).index[0]
        last_ts = int(last_dt.timestamp()*1000)
        if end_dt!=None:
            print(f'USER SELECTED END DATE {end_dt}')
            stop_dt = end_dt
            end_dt =  last_dt  + timedelta(days=7)
            end_ts = int(end_dt.timestamp()*1000)  

        else :
            print('AUTO SHIFT')

            stop_dt = last_dt  + timedelta(days=60)
            end_dt =  last_dt  + timedelta(days=7)
            end_ts = int(end_dt.timestamp()*1000)    
        print(f'{s} Update DATA , Start {last_dt} :  Ending Update {stop_dt}')
        print(f'#'*100) 
        
    while last_dt <stop_dt:
        print('='*100)
        print(f'Update Data {last_dt} :  {end_dt}')
        df1 = binance_ohlc(s,since=int(last_ts),end=end_ts).set_index('date')
        
        
        
        last_dt = df1.tail(1).index[0]
        last_ts = int(last_dt.timestamp()*1000)
        end_dt =  last_dt  + timedelta(days=7)
        end_ts = int(end_dt.timestamp()*1000) 

        df  = pd.concat([df,df1])
        df = df.reset_index().drop_duplicates(subset= 'date',keep='last')
        df = df.set_index('date')
    df = df.loc[(df.index>=start_dt.strftime('%x'))& (df.index<=stop_dt.strftime('%x'))]
    
    return df
```
