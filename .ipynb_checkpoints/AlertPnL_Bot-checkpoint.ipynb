```python
sl_trig =0.05
tp_trig =0.10

# Create DataFrame

def check_Position(trades,sl_trig=0.02,tp_trig=0.01) :
    trade_order =pd.DataFrame(trades)
    trade_order['sl_price'] = 0
    trade_order['tp_price'] = 0
    trade_order['last'] = 0  
    trade_order['diff'] = 0  
    trade_order['diff_p'] = 0  
    trade_order['sl_cond'] = False #  pnl below sl
    trade_order['tp_cond'] = False  # pnl above tp 
    trade_order['exit_cond'] = False # price below sl , price a 
    tp_price = 0 
    sl_price = 0
    # Loop calculation SL,TP,DIFF

    for i,row in trade_order.iterrows():
        
        last = ex.fetch_ticker(row['symbol'] )['last']
        trade_order.loc[i,'last'] = last
        
        if row['side'] =='short':
            trade_order.loc[i,'diff'] = ( row['entry_price'] - last)
            trade_order.loc[i,'diff_p'] = ( row['entry_price'] - last) /row['entry_price']

            if row['sl_mode'] == 'on':
                    sl_price = row['entry_price']+ ( row['entry_price'] *sl_trig )
                    trade_order.loc[i,'sl_price'] = sl_price 

            if row['tp_mode'] =='on':
                    tp_price=  row['entry_price']-(row['entry_price'] * tp_trig)
                    trade_order.loc[i,'tp_price'] = tp_price
                    
            if last > sl_price:
                trade_order.loc[i,'exit_cond'] = True
                
            if last < tp_price :
                trade_order.loc[i,'tp_cond'] = True

        else :
            # Long Position
            trade_order.loc[i,'diff'] = ( last- row['entry_price'])
            trade_order.loc[i,'diff_p'] =( last- row['entry_price']) /row['entry_price']

            if row['sl_mode'] == 'on':
                        sl_price = row['entry_price']- ( row['entry_price'] *sl_trig )
                        trade_order.loc[i,'sl_price'] = sl_price

            if row['tp_mode'] =='on':
                    tp_price=  row['entry_price']+(row['entry_price'] * tp_trig)
                    trade_order.loc[i,'tp_price']=  tp_price
                    
            if last < sl_price:
                trade_order.loc[i,'exit_cond'] = True
                
            if last > tp_price :
                trade_order.loc[i,'tp_cond'] = True
    trade_details = trade_order.copy()
#     trade_details = trade_order[['date','symbol','side','entry_price','unit','sl_price','tp_price','last'\
#                                  ,'diff','diff_p']]
    trade_details['pnl']   =(trade_details['entry_price'] * trade_details['unit'])* trade_details['diff_p']
    return trade_details
```
