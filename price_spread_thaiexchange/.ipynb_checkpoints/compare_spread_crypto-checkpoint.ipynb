```python
## CCXT Function 
def get_data(ex,symbols,**kwarg):
    # 
    
    df= pd.DataFrame(ex.fetch_ohlcv(symbols,**kwarg ),columns=['date','open','high','low','close','volume'])
    df['date']=df['date'].apply(ex.iso8601)
    df['date']=pd.to_datetime(df['date'])
    
    
    #   df['date']=df['date'].apply(lambda x: x.strftime('%Y-%m-%d %H:%M'))
    #   df['date']=pd.to_datetime(df['date'])
    #   df.dropna(inplace=True)

    df.set_index('date',inplace=True)
    return df



def spread_symbol(exchange,symbol):
    res = exchange.fetchOrderBook(symbol,limit=100)
    ts=  dt.now()


    min_data = min(len(res['bids']), len(res['asks']))
    res['bids']  = res['bids'][:min_data]
    res['asks']  = res['asks'][:min_data]
    order_book = pd.DataFrame(res)
    b_size = [order_book['bids'][i][1] for i in range(len(order_book))]
    order_book['bids_size'] = b_size

    b_price =  [order_book['bids'][i][0] for i in range(len(order_book))]
    order_book['bids_price'] = b_price


    a_size = [order_book['asks'][i][1] for i in range(len(order_book))]
    a_price  = [order_book['asks'][i][0] for i in range(len(order_book))]
    order_book['asks_size'] = a_size
    order_book['asks_price'] = a_price


    order_book = order_book[['bids_price','bids_size','asks_price','asks_size']]
    order_book['value_bid'] = order_book['bids_price'] * order_book['bids_size']
    order_book['value_ask'] = order_book['asks_price'] * order_book['asks_size']
    res =exchange.fetch_ticker(symbol)
    last = res['last']
    asks=  res['ask']
    bids = res['bid']
    spread = (asks/bids) -1 # asks/bids -1 , (offer -bid) /offer
    spread_asks = last/asks -1
    spread_bid =bids/last -1
    
    order_book[['cumvalue_bid','cumvalue_ask']] =order_book[['value_bid','value_ask']].cumsum()
    order_book[['cumvalue_bid','cumvalue_ask']] = order_book[['cumvalue_bid','cumvalue_ask']].astype(int)
    order_book['diff_bid']= (order_book['bids_price'] /order_book['bids_price'].iloc[0] -1)*100
    order_book['diff_ask']= (order_book['asks_price'] /order_book['asks_price'].iloc[0] -1)*100
    
    market_bid_chg  = (order_book['bids_price'].iloc[-1]/order_book['bids_price'].iloc[0] -1)*100 # Market Bids
    market_ask_chg  = (order_book['asks_price'].iloc[-1]/order_book['asks_price'].iloc[0] -1)*100 # Market Asks

    
    print(f' [{ts}] [{symbol}]  {bids} : {asks}  Spread @ {spread*100:.2f}%: Last{last}')
    
    print('ALL BID',market_bid_chg,'%',order_book['cumvalue_bid'].iloc[-1],' $') 
    print('ALL ASKS',market_ask_chg,'%',order_book['cumvalue_ask'].iloc[-1],' $')     
    
#     print(f' [{ts}] [{symbol}]  Spread Bid{spread_bid*100:.2f} %   offer{spread_offer*100 :.2f} % ')
    return order_book

# กวาด x bath
def check_ladder(df:pd.DataFrame,cumvalue_col_name:str,trigger:int):
    try:
        f_price =    df['bids_price'].iloc[0]
        dfs = df[df[cumvalue_col_name] >=trigger]
        n_price=  dfs['bids_price'].iloc[0]
        diff_price=(n_price  /f_price -1) 

        print(f'{f_price} : {n_price} Diff {diff_price*100:.2f} %')
        return  dfs
    except IndexError as e:
        print('Over Limit' ,df[cumvalue_col_name].iloc[-1],f'  ')
        return df.iloc[[-1]]
# order_book['cumvalue_bid'].iloc[0]
```

```python
def spread_symbol_bitkub(symbol,limit=100):
    
    ts=  dt.now()
    res= bitkub.books(sym=symbol, lmt=limit)['result']
    min_data = min(len(res['bids']), len(res['asks']))
    res['bids']  = res['bids'][:min_data]
    res['asks']  = res['asks'][:min_data]
    order_book = pd.DataFrame(res)
    order_book['bids_price'] = [order_book['bids'][i][3] for i in range(len(order_book))]
    order_book['bids_size'] = [order_book['bids'][i][4] for i in range(len(order_book))]

    order_book['asks_price'] = [order_book['asks'][i][3] for i in range(len(order_book))]
    order_book['asks_size'] = [order_book['asks'][i][4] for i in range(len(order_book))]
    
    order_book['value_bid'] = [order_book['bids'][i][2] for i in range(len(order_book))]
    order_book['value_ask'] = [order_book['asks'][i][2] for i in range(len(order_book))]
    order_book = order_book.iloc[:,2:]
    # Calculation 
    order_book[['cumvalue_bid','cumvalue_ask']] =order_book[['value_bid','value_ask']].cumsum()
    order_book[['cumvalue_bid','cumvalue_ask']] = order_book[['cumvalue_bid','cumvalue_ask']].astype(int)
    order_book['diff_bid']= (order_book['bids_price'] /order_book['bids_price'].iloc[0] -1)*100
    order_book['diff_ask']= (order_book['asks_price'] /order_book['asks_price'].iloc[0] -1)*100
        
    
    
    re2 = bitkub.ticker(sym=symbol)
    last = re2[symbol]['last']
    asks=  re2[symbol]['lowestAsk']
    bids = re2[symbol]['highestBid']
    spread = (asks/bids) -1 # asks/bids -1 , (offer -bid) /offer
    spread_asks = last/asks -1
    spread_bid =bids/last -1
    print(symbol,bids,asks)

    market_bid_chg  = (order_book['bids_price'].iloc[-1]/order_book['bids_price'].iloc[0] -1)*100 # Market Bids
    market_ask_chg  = (order_book['asks_price'].iloc[-1]/order_book['asks_price'].iloc[0] -1)*100 # Market Asks

    
    print(f' [{ts}] [{symbol}]  {bids} : {asks}  Spread @ {spread*100:.2f}%: Last{last}')
    
    print('ALL BID',market_bid_chg,'%',order_book['cumvalue_bid'].iloc[-1],' $') 
    print('ALL ASKS',market_ask_chg,'%',order_book['cumvalue_ask'].iloc[-1],' $')     
    
#     print(f' [{ts}] [{symbol}]  Spread Bid{spread_bid*100:.2f} %   offer{spread_offer*100 :.2f} % ')
# RequestTimeout
    return order_book
```

```python
def cal_spreard_ccxt(exchange,name_list):

    ''' exchange =ccxt.ExchangeClass
        name_sym ['list'] = symbol name'''
    data = []
    ex_name = exchange.name 
    for name_sym in name_list:
        if ex_name == 'Zipmex':
            
            symbols = name_sym+'/THB'
            
        elif  ex_name == 'Binance' :
            symbols = name_sym+'/USDT'
                    
        elif  ex_name == 'FTX' :
            symbols = name_sym+'/USD'

        res= exchange.fetch_ticker(symbols)
        last = res['last']
        ask = res['ask']
        bid = res['bid']
        volume= res['quoteVolume']
        spread=  ((ask-bid) / ask)*100
        if last >=1 :
            dict_a = {'exchange':ex_name,'symbol':symbols,'name':name_sym,'last':last,'volume':volume,'bid':bid,'ask':ask,'spread':spread}
            data.append(dict_a)
    return data
cal_spreard_ccxt(ftx,['SOL'])
```

```python
def get_price_bk(sym):
    try:
        req = '/api/market/ticker'
        url = API_HOST+req
        syms= 'THB_'+ sym
        param ={'sym':syms}
        res = requests.get(url,params=param).json()
        last = res[syms]['last']
        return last 
    except Exception as e:
        print(str(e), syms)
        
def bk_ticker(symbols='all'): 
    ''' symbols : List'''
    bit_kub_list = []
    all_res = {}

    req = '/api/market/ticker'
    url = API_HOST+req
    if symbols =='all':
        param ={'query':'all'}
        all_res = requests.get(url,params=param).json()


    else :
        for sym in symbols:
#             syms= 'THB_'+ sym
            param ={'sym':sym}
            print(sym)
            res = requests.get(url,params=param).json()
            print(res)
            all_res.update(res)
#     print('Check: ',url,param)
    return all_res



def cal_spread_bk(all_res):
    data = []
    for i in all_res:
        name= i.split('_')[1]
        last = float(all_res[i]['last'])
        volume = float(all_res[i]['quoteVolume'])
        bid = float(all_res[i]['highestBid'])
        ask = float(all_res[i]['lowestAsk'])
        spread = ((ask -bid) /ask)*100

        if last > 1.0:
            dict_a = {'exchange':'bitkub','symbol':i,'name':name,'last':last,'volume':volume,'bid':bid,'ask':ask,'spread':spread}
            data.append(dict_a)
    return data



def cal_spread(ex,name_market,bk=False ):
    data=  []
    for i in tqdm(name_market):
#         detail = zm.fetch_ticker(i)
    # list from bitkub  
        if bk :
            try:
                detail = ex.fetch_ticker(i+'/THB')
            except Exception as e:
                print(str(e))
                detail= None
        else:
            detail = ex.fetch_ticker(i)
        if detail !=None :
            price= detail['last']
            volumex=  detail['quoteVolume']
            namex= detail['symbol'].split('/')[0]
#             bid = float(detail['info']['BestBid']) 
            bid = float(detail['bid']) 
            ask = float(detail['ask']) 

#             ask  = float(detail['info']['BestOffer'])
            try:
                spread = ((ask -bid) /ask)*100
            except  Exception as e:
                print(str(e))
                spread = None
            print(bid,ask,spread,i)


            dict_a  = dict(symbol=i,name=namex,last=price,volume=volumex,bid=bid,ask=ask,spread=spread)
            data.append(dict_a)
            time.sleep(zm.rateLimit/1000)
    return data



```

```python
def compare_spread(ex,spread_list):      
    compare_spread = []
    last_price = None
    
    for i in tqdm(spread_list):
        name = i['name']
        if ex.name=='FTX':
            symbols = i['name']+'/USD'
        elif ex.name =='Zipmex':
            symbols= i['name']+'/THB'

        else:
#             print(ex.name)
            symbols = i['name']+'/USDT'


        try:
            res = ex.fetch_ticker(symbols)
            last_price = res['last']
            bid = ex.fetch_ticker(symbols)['bid']
            ask = ex.fetch_ticker(symbols)['ask']
            spread = ((ask -bid) /ask)*100

        except Exception as e:
            last_price= None 
            spread =None 
            print(str(e))
            
            
        if last_price != None:   
            dict_b= {'exchange':ex.name,'symbol':symbols,'name':name,'last':last_price,'bid':bid,'ask':ask,'spread':spread}
            compare_spread.append(dict_b)
            
    return compare_spread
```

```python
# quick view
name_sym = 'SOL'
usd_thb = get_thb()
def fast_view(name_sym:str,debug=True):
    
    bk_sym = 'THB_'+name_sym
    p1=spread_symbol(ftx,name_sym+'/USDT')
    p2=spread_symbol(binance,name_sym+'/USDT')
    p3= spread_symbol(zm,name_sym+'/THB')

    p4 =  bk_ticker([bk_sym])

    # display(p3.iloc[[0]],p2.iloc[[0]],p1.iloc[[0]])


    zm_usd  =float(round((p3.iloc[[0]]['asks_price'].values[0] / usd_thb),5))
    bk_usd = float(round(p4[bk_sym]['last'] / usd_thb,5))
    bn_usd = p2.iloc[[0]]['asks_price'].values[0]
    ftx_usd = p1.iloc[[0]]['asks_price'].values[0]


    diff = (zm_usd/bn_usd)-1
    diff2 = (zm_usd/ftx_usd)-1
    diff3 = (ftx_usd/bn_usd)-1

    diff4 = (bk_usd/bn_usd)-1
    diff5 = (bk_usd/ftx_usd)-1
    diff6 = (bk_usd/zm_usd)-1
    if debug:
        print(f' Zipmex {zm_usd:.5f} Binance {bn_usd:.5f} {diff*100:.2f}%')
        print(f' Zipmex {zm_usd:.5f} FTX {ftx_usd:.5f}  {diff2*100:.2f}%')
        print(f' FTX {ftx_usd:.5f} Binance {bn_usd:.5f}  {diff3*100:.2f}%')
        print(f' BITKUB {bk_usd:.5f} Binance {bn_usd:.5f}  {diff4*100:.2f}%')
        print(f' BITKUB {bk_usd:.5f} FTX {ftx_usd:.5f}  {diff5*100:.2f}%')
        print(f' BITKUB {bk_usd:.5f} Zipmex {zm_usd:.5f}  {diff6*100:.2f}%')
    dict_data= {'symbol':name_sym,'zipmex':zm_usd,'binance':bn_usd,'ftx':ftx_usd,'bk_usd':bk_usd}
    return pd.DataFrame([dict_data])


all_df = pd.DataFrame()
for  i in ['BNB','ETH','SOL','XRP','LTC','FTT']:
    df=  fast_view(i,debug= False)
    all_df  = all_df.append(df)
```
